### Look at the amount of letters and amount of words in the track name

The issue is, I first need to remove track name components with anything after a -,:, or ().  
This removes all of the "remaster" and features from counting in my letter and word count.  
I have adjusted it so that it only removes stuff at the end of the track name, but this will mess up some like Save a Horse (Ride a Cowboy)

```python
special_cases = df[df['track_name'].str.match(r'.*\([A-Za-z].*\)$')]

# Show only the track_name column and a random 100
print(special_cases['track_name'].sample(100))

print(len(special_cases))
```

So it looks like there are 1065 () cases, and I would have to go through every single on to figure out which ones have important words in the parentheses.  
This sounds very difficult, and these special cases are kind of rare, so i will just ignore them for the purpose of this project :(

```python
#makes new column without anything after a -,:, or (). 
#This removes all of the "remaster" and features from counting in my letter and word count
#only does the for stuff at the end of the string so that it keep words important to the title
    #Ex: 
df.insert(
    2, #insert new column next to track name
    'clean_track_name', #name for new column
    df['track_name'].str.replace(r'\s*(\([^)]*\)|-.*|:.*)$', '', regex=True) 
)
#note for str.replace
    #\s*                 → optional spaces before extra info
    #(\([^)]*\)|-.*|:.*) → match either: parentheses with anything inside, a dash + anything, or a colon + anything
    #$                   → only at the end of the string

#sum letters in track_name string excluding spaces, special characters, and numbers; then make new column
df.insert(3, 'letters_track_name', df['clean_track_name'].str.count(r'[A-Za-z]'))

df.head(100) #show me more rows for checking
```

### Run a Multiple Linear Regression to see which variables might affect song popularity!

```python
import statsmodels.formula.api as smf

model = smf.ols(
    'popularity ~ numeric1 + numeric2 + C(cat1) + C(cat2)',
    data=df
).fit()

print(model.summary())
```
